```python
trainX = []
trainY = []

with open('../IMDB/imdb_trainX.txt' , 'r') as f:
    Lines = f.readlines()
    for line in Lines:
        trainX.append(line)
    
with open('../IMDB/imdb_trainY01.txt' , 'r') as f:
    Lines = f.readlines()
    for line in Lines:
        trainY.append(line)
    
print(len(trainX))
print(len(trainY))
```

```python
import re
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import numpy as np

tokenizer = RegexpTokenizer(r'\w+')
stopwordsInEnglish = set(stopwords.words('english'))
stm = PorterStemmer()

def getCleanReview(review):
    review = review.lower()
    review = review.replace("<br /><br />"," ")
    
    tokens = tokenizer.tokenize(review)
    new_tokens = [token for token in tokens if token not in stopwordsInEnglish]
    stemmed_tokens = [stm.stem(token) for token in new_tokens]
    
    cleaned_review = ' '.join(stemmed_tokens)
    
    return cleaned_review
```

```python
dicWords = {}
```

```python
cnt = 0
for review in trainX:
    reviewClean = getCleanReview(review)
    reviewSplit = reviewClean.split()
    for word in reviewSplit:
        negative = 0
        positive = 0
        
        if dicWords.get(word)== None:
            if trainY[cnt] == "0\n":
                negative += 1
            else:
                positive += 1
                
            dicWords[word] = [negative , positive]
        else:
            negative = dicWords[word][0]
            positive = dicWords[word][1]
            if trainY[cnt] == "0\n":
                negative += 1
            else:
                positive += 1
                
            dicWords[word] = [negative , positive]
    cnt+=1
```

```python
print(len(dicWords))
```

```python
neg_vocab_count = 0
pos_vocab_count = 0
neg_vocab_size = 0
pos_vocab_size = 0
for key,value in dicWords.items():
    if value[0]!=0:
        neg_vocab_count += value[0]
        neg_vocab_size +=1
    if value[1]!= 0:
        pos_vocab_count += value[1]
        pos_vocab_size +=1
```

```python
print(neg_vocab_size)
print(pos_vocab_size)
print(neg_vocab_count)
print(pos_vocab_count)
```

```python
class trainNB:
    def __init__(self , dicWords , neg_vocab_count , pos_vocab_count , neg_vocab_size , pos_vocab_size):
        self.dicWords = dicWords
        self.neg_vocab_count = neg_vocab_count
        self.pos_vocab_count = pos_vocab_count
        self.neg_vocab_size = neg_vocab_size
        self.pos_vocab_size = pos_vocab_size

trainObj = trainNB(dicWords = dicWords , neg_vocab_count = neg_vocab_count , pos_vocab_count = pos_vocab_count , neg_vocab_size = neg_vocab_size , pos_vocab_size = pos_vocab_size)
```

```python
import pickle as pk
pickle_out = open("trainDicNB.pickle","wb")
pk.dump(trainObj,pickle_out)
pickle_out.close()
```
