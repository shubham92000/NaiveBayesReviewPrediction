```python
class trainNB:
    def __init__(self , dicWords , neg_vocab_count , pos_vocab_count , neg_vocab_size , pos_vocab_size):
        self.dicWords = dicWords
        self.neg_vocab_count = neg_vocab_count
        self.pos_vocab_count = pos_vocab_count
        self.neg_vocab_size = neg_vocab_size
        self.pos_vocab_size = pos_vocab_size
```

```python
import pickle as pk
pickle_in = open("trainDicNB.pickle","rb")
NBclassifier = pk.load(pickle_in)
```

```python
print(len(NBclassifier.dicWords))
```

```python
import re
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import numpy as np

tokenizer = RegexpTokenizer(r'\w+')
stopwordsInEnglish = set(stopwords.words('english'))
stm = PorterStemmer()

def getCleanReview(review):
    review = review.lower()
    review = review.replace("<br /><br />"," ")
    
    tokens = tokenizer.tokenize(review)
    new_tokens = [token for token in tokens if token not in stopwordsInEnglish]
    stemmed_tokens = [stm.stem(token) for token in new_tokens]
    
    cleaned_review = ' '.join(stemmed_tokens)
    
    return cleaned_review
```

```python
testX = []
testY = []
    
with open('../IMDB/imdb_testX.txt' , 'r') as f:
    Lines = f.readlines()
    for line in Lines:
        testX.append(line)
    
with open('../IMDB/imdb_testY01.txt' , 'r') as f:
    Lines = f.readlines()
    for line in Lines:
        testY.append(line)
    
print(len(testX))
print(len(testY))
```

```python
y_pred = [None]*len(testX)
```

```python
print(len(y_pred))
```

```python
alpha = 1
count = 0
for review in testX:
    cleantest = getCleanReview(review)
    clean_test_words = cleantest.split()
    prob_nega = 0
    prob_posi = 0
    for word in clean_test_words:
        if NBclassifier.dicWords.get(word) != None:
            cnt = NBclassifier.dicWords[word]
        else:
            cnt = [0,0]

        prob_nega = prob_nega+np.log((cnt[0] + alpha)/(NBclassifier.neg_vocab_count + alpha*NBclassifier.neg_vocab_size))
        prob_posi = prob_posi+np.log((cnt[1] + alpha)/(NBclassifier.pos_vocab_count + alpha*NBclassifier.pos_vocab_size))

    if prob_nega > prob_posi:
        y_pred[count] = 0
    else:
        y_pred[count] = 1
    
    count+=1
```

```python
for e in y_pred:
    print(e)
```

```python
print(len(y_pred))
```

```python
correctly_classified_Points = 0    #accuracy
for i in range(len(y_pred)):
    if y_pred[i] == int(testY[i]):
        correctly_classified_Points += 1
        
print(correctly_classified_Points)
print(correctly_classified_Points/len(y_pred))
```
